# Gradient descent #



```python
import matplotlib
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from matplotlib import cm, lines, collections, colors
```

```python
# Define the function

delta = 0.01 # spacing between samples in each dimension
x = np.arange(-5.0, 5.0, delta)
y = np.arange(-5.0, 5.0, delta)
X, Y = np.meshgrid(x, y)

epsilon_arr = [1,0.1,0.01,0.001]
epsilon = epsilon_arr[0]
Z = X**2+2*(1-epsilon)/(1+epsilon)*X*Y+Y**2 # f(x,y) for all x,y pairs

```

```python
# Plot the functions
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122)


# Surface plot
surf = ax1.plot_surface(X, Y, Z, cmap=cm.RdGy, linewidth=0, antialiased=False)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=5)

# Contour plot
contours = ax2.contour(X, Y, Z, 20)
ax2.clabel(contours, inline=True, fontsize=10)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
plt.show()
```

```python
# Define the gradient descent algorithm
def _func(epsilon):
    # Starting point
    x_0 = np.array([1.0, 1.0])

    # Gradient
    def gradient(v):
        # the 1st dimension is x, the 2nd dimension is y
        #grad = np.array([2*a*v[0], 2*v[1]])
        #grad = np.array([[2*v[0]+2*v[1]*(1-epsilon)/(1+epsilon)],
        #                 [2*v[0]*(1-epsilon)/(1+epsilon)+2*v[1]]])
        grad = np.array([2*v[0]+2*v[1]*(1-epsilon)/(1+epsilon),
                         2*v[0]*(1-epsilon)/(1+epsilon)+2*v[1]])
        return grad
    
    ## Evaluate function
    def func_eval(v):
        x,y = v[0],v[1]
        val = x**2+2*(1-epsilon)/(1+epsilon)*x*y+y**2
        return val

    # Step-size
    #alpha = 0.01
    m = 2-2*(1-epsilon)/(1+epsilon)
    M = 2+2*(1-epsilon)/(1+epsilon)
    alpha = m/2/M**2
    # Tolerance
    tolerance = 1e-6

    # Gradient descent iteration
    x_arr = []
    x_arr += [list(x_0)]
    x = x_0
    lines = []
    val = []
    val += [func_eval(x_0)]
    for ii in range(1000):   
        step = -alpha * gradient(x)
        if np.all(np.abs(step) <= tolerance):
            break
        x_prime = x + step
        x_arr += [list(x_prime)]
        val += [func_eval(x_prime)]
        lines.append([x.tolist(), x_prime.tolist()])
        x = x_prime
    
    print(f'Final iterate = {x}.')
    print(f'Number of iterations = {ii}.')
    return np.array(x_arr),lines,val
```

```python
x,lines,fx = _func(epsilon_arr[0]) ##Get the solution with different epsilon
```

```python
plt.figure(figsize=(15,5))
plt.suptitle('Gradient-descent with $\epsilon$=1',fontsize=20)
plt.subplot(1,2,1)
plt.scatter(x[1:-1,0],x[1:-1,1])
plt.scatter(x[0,0],x[0,1],marker = 'x',label='initial')
plt.scatter(x[-1,0],x[-1,1],marker='^',label='final')
plt.title('scatter plot for each iteration')
plt.legend()
plt.subplot(1,2,2)
plt.plot(fx)
plt.title('$f(x_1,x_2)$ for each iteration')
plt.xlabel('number of iteartion n')
#plt.xticks(np.arange(0,len(c),2))
```

## Backtracking_Line_Search

```python
# Define the gradient descent algorithm
def _func(epsilon):
    # Starting point
    x_0 = np.array([1.0, 1.0])

    # Gradient
    def gradient(v):
        # the 1st dimension is x, the 2nd dimension is y
        #grad = np.array([2*a*v[0], 2*v[1]])
        #grad = np.array([[2*v[0]+2*v[1]*(1-epsilon)/(1+epsilon)],
        #                 [2*v[0]*(1-epsilon)/(1+epsilon)+2*v[1]]])
        grad = np.array([2*v[0]+2*v[1]*(1-epsilon)/(1+epsilon),
                         2*v[0]*(1-epsilon)/(1+epsilon)+2*v[1]])
        return grad
    def func_eval(v):
        x,y = v[0],v[1]
        val = x**2+2*(1-epsilon)/(1+epsilon)*x*y+y**2
        return val
    
    def get_a(v,beta):
        a = 1.0
        while func_eval(v-a*gradient(v))>func_eval(v) - a/2*(gradient(v)@gradient(v)):
            a *= beta
        return a
    # Step-size
    #alpha = 0.01
    #m = 2-2*(1-epsilon)/(1+epsilon)
    #M = 2+2*(1-epsilon)/(1+epsilon)
    #alpha = m/2/M**2
    # Tolerance
    tolerance = 1e-6

    # Gradient descent iteration
    x_arr = []
    x_arr += [list(x_0)]
    x = x_0
    lines = []
    val = []
    val += [func_eval(x_0)]
    alpha_arr = []
    for ii in range(1000): 
        alpha = get_a(x,1/np.sqrt(2))
        alpha_arr += [alpha]
        step = -alpha * gradient(x)
        if np.all(np.abs(step) <= tolerance):
            break
        x_prime = x + step
        x_arr += [list(x_prime)]
        val += [func_eval(x_prime)]
        lines.append([x.tolist(), x_prime.tolist()])
        x = x_prime
    
    print(f'Final iterate = {x}.')
    print(f'Number of iterations = {ii}.')
    return alpha_arr,np.array(x_arr),lines,val
```

```python
a,x,lines,fx = _func(epsilon_arr[1])
plt.figure(figsize=(15,5))
plt.suptitle('Gradient-descent with $\epsilon$=0.1',fontsize=20)
plt.subplot(1,2,1)
plt.scatter(x[1:-1,0],x[1:-1,1])
plt.scatter(x[0,0],x[0,1],marker = 'x',label='initial')
plt.scatter(x[-1,0],x[-1,1],marker='^',label='final')
plt.title('scatter plot for each iteration')
plt.legend()
plt.subplot(1,2,2)
plt.plot(fx)
plt.title('$f(x_1,x_2)$ for each iteration')
plt.xlabel('number of iteartion n')
#plt.xticks(np.arange(0,len(c),2))
```
