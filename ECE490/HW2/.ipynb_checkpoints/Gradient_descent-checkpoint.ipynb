# Gradient descent #



```python
import matplotlib
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from matplotlib import cm, lines, collections, colors
```

Let's code gradient descent on a quadratic example function $f(x,y) = a x^2 + y^2$ with $0 < a < 1$.

```python
# Define the function

delta = 0.01 # spacing between samples in each dimension
x = np.arange(-5.0, 5.0, delta)
y = np.arange(-5.0, 5.0, delta)
X, Y = np.meshgrid(x, y)

a = 1/2
Z = a * (X**2) + Y**2 # f(x,y) for all x,y pairs

Q = np.array([[2*a, 0],[0, 2]]) # Hessian
m = 2*a # smallest eigenvalue
M = 2   # largest eigenvalue
```

Notice that $f$ is convex quadratic with $\nabla^2 f = \begin{pmatrix} 2a & 0 \\ 0 & 2\end{pmatrix}$. 
Call this matrix $Q$. 

Eigenvalues of $Q$ are $2a , 2$. 

What is the Lischitz constant for $\nabla f$? Show that it is $M = 2$. Therefore, gradient descent with constant step-size will converge if the step size is less than $2/M = 1$. 

The condition number is $\kappa = 1/a$. As you reduce $a$, $f$ near the minimum looks like a *valley* that impedes convergence. Try reducing $a$ from $1/2$ to $1/20$. Fix some step-size and check what happens to the number of iterations.

```python
# Plot the functions
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122)


# Surface plot
surf = ax1.plot_surface(X, Y, Z, cmap=cm.RdGy, linewidth=0, antialiased=False)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=5)

# Contour plot
contours = ax2.contour(X, Y, Z, 20)
ax2.clabel(contours, inline=True, fontsize=10)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
plt.show()
```

Let's code up gradient descent with constant step-size. Check what happens with step-sizes of $1.0, 0.3, 0.01$.
Notice the gradient is $\nabla f(x,y) = \begin{pmatrix} 2a x \\ 2y \end{pmatrix}$

```python
# Define the gradient descent algorithm

# Starting point
x_0 = np.array([3.0, 3.0])

# Gradient
def gradient(v):
    # the 1st dimension is x, the 2nd dimension is y
    grad = np.array([2*a*v[0], 2*v[1]])
    return grad

# Step-size
alpha = 0.01

# Tolerance
tolerance = 1e-8

# Gradient descent iteration
x = x_0
lines = []
for ii in range(1000):   
    step = -alpha * gradient(x)
    if np.all(np.abs(step) <= tolerance):
        break
    x_prime = x + step
    lines.append([x.tolist(), x_prime.tolist()])
    x = x_prime
    
print(f'Final iterate = {x}.')
print(f'Number of iterations = {ii}.')
```

```python
# Contour plot again with descent dynamics
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot()

contours = ax.contour(X, Y, Z, 50)

# test a plain-ol-line
# lines = [[x.tolist(), (1,1)], [[1,1], [-1,0]]]
lc = collections.LineCollection(lines,linewidths=2)
ax.add_collection(lc)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```
