```python
dv = 0.1
N = 10000
v = np.linspace(-10,10,200,endpoint = False)
#v = -8 + np.arange(160)*dv
t = np.linspace(0,100,N,endpoint=False)
f1 = 0.1*e**(-(v-1)**2/2)
df0 = -v * 1/sqrt(2*pi)*e**(-v**2/2)
E = np.zeros(N,np.complex64)
#E[0] = 0.01
```

```python
k = 0.2
for i in range(1,(N)):
    f1 = f1 + (t[1]-t[0]) * (-1j * k * v * f1 + E[i-1] * df0)
    E[i] = 1j/k*sum(f1) * dv
    #print(i)
```

```python
plot(t,real(E))
xlabel('time in [s]')
ylabel('Re{E}')
```
